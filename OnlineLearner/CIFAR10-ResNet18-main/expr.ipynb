```python
import numpy as np
import torchvision as tv

ds = tv.datasets.FashionMNIST(root='data/fashion-mnist', train=True, download=True)
print(ds.classes)
print(len(ds))
```

```python
images = np.zeros([len(ds), *(ds[0][0].size)])
print(images.shape)
for i in range(len(ds)):
    images[i, :, :] = np.array(ds[i][0])
```

```python
import matplotlib.pyplot as plt
plt.imshow(images[0], cmap='jet')
```

```python
mean_ = images.mean()/255.0
print(mean_)
std_ = images.std()/255.0
print(std_)
```

```python
# for cifar10
cifar10 = tv.datasets.CIFAR10(root='data/cifar-10-python', train=True, download=False)
print(cifar10.classes)
print(len(cifar10))
```

```python
images = np.zeros([len(cifar10), *(cifar10[0][0].size), 3])
print(images.shape)
for i in range(len(cifar10)):
    images[i, :, :, :] = np.array(cifar10[i][0])
```

```python
print(images[0].dtype)
plt.imshow(np.uint8(images[5]))
```

```python
mean_ = images.mean(axis=(0,1,2))/255.0
print(mean_)
std_ = images.std(axis=(0,1,2))/255.0
print(std_)
```
